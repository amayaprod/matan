**Условие:**

Докажите, что для любого $n \in \mathbb{N}$, $r \in [0; 1]$
$$0 < x_0 < 1 \implies 0 < x_n < 1$$

---

**Доказательство (методом математической индукции):**

**1. База индукции ($n=0$):**
По условию задачи $0 < x_0 < 1$. Утверждение верно.

**2. Индукционный переход:**
Предположим, что для некоторого шага $k$ утверждение верно: $x_k \in (0; 1)$. Докажем, что тогда и $x_{k+1} \in (0; 1)$.

Рассмотрим функцию отображения $f(x) = r x (1 - x) (2 - x)$ на интервале $x \in (0; 1)$.

* **Оценка снизу ($x_{k+1} > 0$):**
Так как $0 < x_k < 1$, то:
$$x_k > 0$$
$$(1 - x_k) > 0$$
$$(2 - x_k) > 1 > 0$$
Поскольку $r > 0$ (рассматриваем нетривиальный случай), произведение положительных чисел дает положительный результат:
$$x_{k+1} = r \cdot x_k \cdot (1 - x_k) \cdot (2 - x_k) > 0$$

* **Оценка сверху ($x_{k+1} < 1$):**
Найдем максимум функции $h(x) = x(1-x)(2-x)$ на отрезке $[0; 1]$.
Раскроем скобки: $h(x) = x^3 - 3x^2 + 2x$.
Найдем экстремум, взяв производную:
$$h'(x) = 3x^2 - 6x + 2 = 0$$
Корни уравнения: $x_{1,2} = \frac{6 \pm \sqrt{36 - 24}}{6} = 1 \pm \frac{\sqrt{3}}{3}$.
На интервале $(0; 1)$ лежит только корень $x_{max} = 1 - \frac{\sqrt{3}}{3} \approx 0.423$.
    
Значение функции в максимуме:
$$h(x_{max}) \approx 0.385$$
    
Тогда для следующего шага имеем:
$$x_{k+1} = r \cdot h(x_k) \le r \cdot h(x_{max}) \approx r \cdot 0.385$$
    
Так как по условию $r \in [0; 1]$, то:
$$x_{k+1} \le 1 \cdot 0.385 < 1$$

**Вывод:**
Из того, что $0 < x_k < 1$, следует, что $0 < x_{k+1} < 1$. Согласно принципу математической индукции, утверждение верно для любого n.

---


### 1. Найдите все неподвижные точки логистического отображения.

$x_{n+1} = r x_n (1 - x_n)$.

### **Решение:**
Неподвижная точка удовлетворяет уравнению $f(x^*) = x^*$.
$$r x (1 - x) = x$$
$$rx - rx^2 - x = 0 \implies x(r - 1 - rx) = 0$$
Корни уравнения:
1.  **$x_1^* = 0$** (существует всегда).
2.  $r - 1 - rx = 0 \implies$ **$x_2^* = 1 - \frac{1}{r}$**.

### 2. При каких r отображение имеет одну неподвижную точку? Несколько?

* **Одна точка ($x^*=0$):** Когда вторая точка $x_2^* \le 0$, то есть при $1 - 1/r \le 0 \implies r \le 1$.
* **Две точки:** Когда $r > 1$, появляется положительная точка $x_2^*$.

### 3. Какое максимальное количество неподвижных точек может иметь логистическое отображение? Почему?
* **Максимальное количество:** **2**.
*Почему?* Уравнение $rx(1-x)=x$ сводится к квадратному уравнению $rx^2 + (1-r)x = 0$. Квадратное уравнение не может иметь более двух действительных корней.

№2

**Условие:**

Докажите, что при $x_0 \in (0; 1)$ и $r \in (0; 1]$ последовательность $\{x_n\}$, заданная логистическим отображением, монотонно убывает. Существует ли предел у данной последовательности при $r \in (0; 1]$? Докажите. Покажите графически.


---


**Доказательство:**
Рассмотрим отношение следующего члена к предыдущему:
$$\frac{x_{n+1}}{x_n} = \frac{r x_n (1 - x_n)}{x_n} = r(1 - x_n)$$
Так как $x_n \in (0; 1)$, то множитель $(1 - x_n) < 1$.
Учитывая, что по условию $r \le 1$, получаем:
$$r(1 - x_n) < 1 \implies \frac{x_{n+1}}{x_n} < 1 \implies x_{n+1} < x_n$$
**Вывод:** Последовательность монотонно убывает. Так как она ограничена снизу нулем ($x_n > 0$), предел существует и равен 0.

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map_series(r, x0, n_steps):
    x = np.zeros(n_steps)
    x[0] = x0
    for i in range(1, n_steps):
        x[i] = r * x[i-1] * (1 - x[i-1])
    return x


steps = 50
x0 = 0.5

r_values = [0.2, 0.5, 0.9, 1.0]

plt.figure(figsize=(10, 6))

for r in r_values:
    sequence = logistic_map_series(r, x0, steps)
    plt.plot(sequence, 'o-', label=f'r = {r}')

plt.title(r'Монотонное убывание последовательности при $r \in (0; 1]$')
plt.xlabel(r'Номер итерации $n$')
plt.ylabel(r'Значение $x_n$')

plt.axhline(0, color='black', linestyle='--', linewidth=1, label='Предел x*=0')
plt.legend()
plt.grid(True)
plt.show()
```

№3

**Условие:**
Пусть $r \in (2; 3)$, $x_{2n} > x^*$, $x_{2n+1} < x^*$.
Что вы можете сказать о монотонности подпоследовательностей $\{x_{2n}\}$ и $\{x_{2n+1}\}$? Докажите. Проверьте графически.

---

### **Решение**

#### **1. Неподвижная точка**
Для логистического отображения $f(x) = rx(1-x)$ при $r > 1$ существует устойчивая неподвижная точка:
$$x^* = 1 - \frac{1}{r}$$

#### **2. Характер сходимости**
Рассмотрим производную функции в неподвижной точке:
$$f'(x) = r(1 - 2x)$$
В точке равновесия:
$$f'(x^*) = r\left(1 - 2\left(1 - \frac{1}{r}\right)\right) = 2 - r$$

Так как по условию $r \in (2; 3)$, то производная лежит в интервале:
$$-1 < f'(x^*) < 0$$
Отрицательная производная означает, что сходимость **колебательная** («паутина» закручивается вокруг точки). Значения чередуются: больше $x^*$, меньше $x^*$.

#### **3. Анализ подпоследовательностей (отображение через шаг)**
Чтобы понять поведение $\{x_{2n}\}$ и $\{x_{2n+1}\}$, рассмотрим отображение $g(x) = f(f(x))$.
Его производная в неподвижной точке:
$$g'(x^*) = (f'(x^*))^2 = (2 - r)^2$$

Так как $-1 < 2-r < 0$, то квадрат этого числа:
$$0 < g'(x^*) < 1$$

Положительная производная $0 < g'(x^*) < 1$ означает, что функция $g(x)$ притягивает траектории к $x^*$ **монотонно** (без скачков вокруг точки).

#### **4. Вывод**

**А) Для четных членов $\{x_{2n}\}$**
По условию мы находимся выше точки равновесия ($x_{2n} > x^*$). Так как производная отображения "через шаг" положительна, траектория притягивается к $x^*$ с той же стороны. Следующая точка $x_{2n+2}$ будет ближе к $x^*$, но все еще выше него:
$x^* < x_{2n+2} < x_{2n}$.

$\Rightarrow$ Подпоследовательность $\{x_{2n}\}$ **монотонно убывает**.

**Б) Для нечетных членов $\{x_{2n+1}\}$**
По условию мы находимся ниже точки равновесия ($x_{2n+1} < x^*$). Следующая нечетная точка $x_{2n+3}$ будет ближе к $x^*$, но все еще ниже него:
$x_{2n+1} < x_{2n+3} < x^*$.

$\Rightarrow$ Подпоследовательность $\{x_{2n+1}\}$ **монотонно возрастает**.

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    return r * x * (1 - x)


r = 2.8
x0 = 0.1
n_iterations = 25

x = np.zeros(n_iterations)
x[0] = x0
for i in range(1, n_iterations):
    x[i] = logistic_map(x[i-1], r)

x_star = 1 - 1/r

indices = np.arange(n_iterations)
even_idx = indices[::2] # 0, 2, 4...
odd_idx = indices[1::2] # 1, 3, 5...

plt.figure(figsize=(10, 6))

plt.plot(indices, x, 'k-', alpha=0.2, label='Траектория')
plt.axhline(x_star, color='green', linestyle='--', label=f'Неподвижная точка x*={x_star:.2f}')

plt.plot(even_idx, x[even_idx], 'ro-', label='Четные (x_2n) -> убывают к x*')

plt.plot(odd_idx, x[odd_idx], 'bo-', label='Нечетные (x_2n+1) -> возрастают к x*')

plt.title(fr'Сходимость подпоследовательностей при $r={r}$')
plt.xlabel('Номер итерации n')
plt.ylabel('Значение $x_n$')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

**Условие:**

Для отображения $g(x_n)$, заданного вариантом:
1. Аналитически найдите неподвижную точку.
2. Найдите или оцените диапазон параметра r, при котором последовательность монотонно сходится к нулю.
3. Постройте графики зависимости $x_n$ от n для нескольких различных значений параметра r.

---

### **Решение**

#### **1. Аналитический поиск неподвижной точки**
Неподвижная точка удовлетворяет условию $x^* = g(x^*)$.
$$x = r x (1 - x) (3 - x)$$

Перенесем все в одну сторону:
$$x - r x (1 - x) (3 - x) = 0$$
$$x \cdot [1 - r(1 - x)(3 - x)] = 0$$

Отсюда получаем:
1.  **Тривиальная неподвижная точка:** $x^*_1 = 0$.
2.  **Нетривиальные точки** находятся из квадратного уравнения $1 - r(3 - 4x + x^2) = 0$. Их наличие зависит от $r$, но для анализа сходимости к нулю (вымирания) нас интересует поведение именно вокруг $x^* = 0$.

#### **2. Диапазон сходимости к нулю**
Чтобы последовательность монотонно сходилась к $x^* = 0$, эта точка должна быть **асимптотически устойчивой**, а производная в ней должна быть положительной и меньше 1.

Найдем производную функции $g(x) = r(3x - 4x^2 + x^3)$:
$$g'(x) = r(3 - 8x + 3x^2)$$

Вычислим значение производной в точке $x^* = 0$:
$$g'(0) = r(3 - 0 + 0) = 3r$$

**Условия монотонной сходимости:**
$$0 < g'(0) < 1$$
$$0 < 3r < 1$$

Отсюда получаем искомый диапазон параметра:
$$r \in \left(0; \frac{1}{3}\right)$$

**Вывод:** При $r < 1/3$ популяция будет монотонно вымирать. При $r > 1/3$ точка $x=0$ теряет устойчивость, и популяция начинает расти.

```python
import numpy as np
import matplotlib.pyplot as plt

ISU = 501983
VARIANT_N = ISU % 5  # 3

def variant_map(x, r):
    return r * x * (1 - x) * (3 - x)

def generate_trajectory(r, x0, steps):
    x = np.zeros(steps)
    x[0] = x0
    for i in range(1, steps):
        val = variant_map(x[i-1], r)
        if abs(val) > 10: val = 10
        x[i] = val
    return x

steps = 30
x0 = 0.1

r_values = [0.2, 0.3, 0.35, 0.5]

plt.figure(figsize=(10, 6))

for r in r_values:
    traj = generate_trajectory(r, x0, steps)
    plt.plot(traj, 'o-', label=f'r = {r}')

plt.title(f'Динамика системы для ISU {ISU} (Вариант N={VARIANT_N})')
plt.xlabel('Номер итерации n')
plt.ylabel('Значение x_n')
plt.axhline(0, color='black', linestyle='--', linewidth=1, label='x*=0')

plt.axhline(0, color='red', alpha=0, label='Порог r = 1/3')

plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Вариант N: {VARIANT_N}")
print(f"Критическое значение r (теория): 1/3 ≈ {1/3:.4f}")
print("На графике видно:")
print(" - При r < 0.33 линия идет к 0.")
print(" - При r > 0.33 линия уходит вверх от 0.")
```

# Hard

№1

**Условие:**
1. Положим $r_{\infty} \approx 3.5699456...$ . Как изменяется длина цикла при изменении параметра $r$ в диапазоне $(3; r_{\infty})$?
2. Для $r \in (3; r_{\infty})$ экспериментально установите, какие ограничения (закономерности) действуют на длину цикла $m$.

---

### **Решение**

#### **1. Изменение длины цикла**
При увеличении параметра $r$ от 3 до критического значения $r_{\infty}$ система претерпевает серию изменений режима, называемых **бифуркациями удвоения периода**:

* **$r \in (1; 3)$:** Существует одна устойчивая неподвижная точка (цикл периода 1).
* **$r = 3$:** Происходит первая бифуркация. Неподвижная точка теряет устойчивость, рождается устойчивый **цикл периода 2** (система прыгает между двумя значениями).
* **$r \approx 3.449$:** Цикл периода 2 теряет устойчивость, рождается устойчивый **цикл периода 4**.
* **$r \approx 3.544$:** Рождается **цикл периода 8**.
* Далее интервалы между бифуркациями стремительно сокращаются, и рождаются циклы периодов 16, 32, 64 и т.д.

#### **2. Закономерность (ограничение на $m$)**
В диапазоне $r \in (3; r_{\infty})$ длина устойчивого цикла $m$ может принимать только значения, являющиеся **степенями двойки**:
$$m = 2^k, \quad \text{где } k \in \mathbb{N}$$
$$m \in \{2, 4, 8, 16, 32, \dots\}$$

Этот процесс называется **сценарием Фейгенбаума** (переход к хаосу через удвоение периода).

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    return r * x * (1 - x)

def get_attractor(r, x0, n_skip=1000, n_show=100):
    x = x0
    for _ in range(n_skip):
        x = logistic_map(x, r)

    trajectory = []
    for _ in range(n_show):
        x = logistic_map(x, r)
        trajectory.append(x)

    return np.unique(np.round(trajectory, 4))

r_values = [2.8, 3.2, 3.5, 3.56]
x0 = 0.2

print(f"{'r':<10} | {'Длина цикла m':<15} | {'Значения цикла'}")
print("-" * 60)

plt.figure(figsize=(12, 6))

for r in r_values:
    cycle_points = get_attractor(r, x0)
    m = len(cycle_points)

    points_str = str(cycle_points) if len(cycle_points) <= 8 else "[...много точек...]"
    print(f"{r:<10} | {m:<15} | {points_str}")

    traj = []
    x = x0
    for _ in range(100): x = logistic_map(x, r)
    for _ in range(50):
        traj.append(x)
        x = logistic_map(x, r)

    plt.plot(traj, 'o-', label=f'r={r} (m={m})')

plt.title('Демонстрация удвоения периода (Hard Level)')
plt.xlabel('Номер итерации n')
plt.ylabel('Значение x_n')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

#### **1. Алгоритм построения**
Лестница Ламерея строится следующим образом:
1. Выбираем начальную точку $x_0$ на оси абсцисс.
2. Проводим вертикальную линию до пересечения с графиком функции $y = f(x)$. Получаем точку $(x_0, x_1)$.
3. Проводим горизонтальную линию до пересечения с биссектрисой $y = x$. Получаем точку $(x_1, x_1)$.
4. Из этой точки снова проводим вертикальную линию до графика функции... и так далее.

#### **2. Вид циклов на графике**
Визуализация позволяет легко определить режим работы системы:

* **Устойчивая неподвижная точка (сходимость):**
    Линия образует «спираль» или «лесенку», которая закручивается внутрь к точке пересечения параболы и диагонали ($x^*$). Шаги становятся всё меньше.

* **Цикл порядка 2 ($m=2$):**
    Траектория замыкается в **квадрат** (или прямоугольник) вокруг неустойчивой неподвижной точки. «Паутина» бегает по вершинам этого квадрата бесконечно.

* **Цикл порядка 4 и выше:**
    Траектория образует сложную замкнутую фигуру (многоугольник), вершины которой лежат на параболе. Количество вершин на параболе соответствует порядку цикла $m$.

* **Хаос:**
    Линия хаотично заполняет всю область между границами, никогда не замыкаясь и не повторяясь.

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    return r * x * (1 - x)

def plot_cobweb(r, x0, n_iter=100, ax=None):
    x = np.linspace(0, 1, 500)
    y = logistic_map(x, r)

    if ax is None:
        fig, ax = plt.subplots(figsize=(6, 6))

    ax.plot(x, y, 'k', linewidth=1.5, label=f'f(x), r={r}')
    ax.plot([0, 1], [0, 1], 'r--', linewidth=1, label='y=x')

    px, py = [x0], [0]

    x_curr = x0
    for _ in range(n_iter):
        x_next = logistic_map(x_curr, r)
        px.append(x_curr)
        py.append(x_next)

        px.append(x_next)
        py.append(x_next)

        x_curr = x_next

    ax.plot(px, py, 'b-', alpha=0.5, linewidth=0.8)

    ax.set_title(f"Лестница Ламерея: r = {r}")
    ax.set_xlabel('$x_n$')
    ax.set_ylabel('$x_{n+1}$')
    ax.grid(True, alpha=0.3)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))

plot_cobweb(r=2.8, x0=0.1, n_iter=50, ax=axes[0, 0])
axes[0, 0].text(0.1, 0.8, "Сходимость (спираль)", fontsize=12, color='green')

plot_cobweb(r=3.2, x0=0.1, n_iter=50, ax=axes[0, 1])
axes[0, 1].text(0.1, 0.8, "Цикл порядка 2 (квадрат)", fontsize=12, color='green')

plot_cobweb(r=3.5, x0=0.1, n_iter=100, ax=axes[1, 0])
axes[1, 0].text(0.1, 0.8, "Цикл порядка 4", fontsize=12, color='green')

plot_cobweb(r=3.9, x0=0.1, n_iter=200, ax=axes[1, 1])
axes[1, 1].text(0.1, 0.8, "Хаос (заполнение)", fontsize=12, color='green')

plt.tight_layout()
plt.show()
```

### **Анализ и Выводы**

Для визуализации изменения режимы работы системы мы построим **бифуркационную диаграмму**.
* По оси X откладывается параметр $r$.
* По оси Y — значения $x_n$, которые принимает система после установления режима (аттрактор).

**Ожидаемое поведение:**
Как и логистическое отображение, данная кубическая функция должна демонстрировать **сценарий удвоения периода (сценарий Фейгенбаума)**:
1.  Сначала одна линия (период 1, устойчивая точка).
2.  Затем линия раздваивается (бифуркация, период 2).
3.  Затем каждая ветвь снова раздваивается (период 4) и так далее.
4.  В конце наступает хаос (закрашенные области).
5.  Внутри хаоса могут наблюдаться "окна прозрачности" (устойчивые циклы периода 3, 5 и т.д.), что является признаком универсальности данного пути к хаосу.

**Ответ на вопрос задачи:**
Да, поведение системы **качественно совпадает** с поведением логистического отображения. Несмотря на другую формулу (наличие кубического члена), механизм перехода к хаосу через каскад бифуркаций удвоения периода сохраняется.

```python
import numpy as np
import matplotlib.pyplot as plt

def variant_map(x, r):
    return r * x * (1 - x) * (3 - x)


r_min, r_max = 0.3, 1.6
n_r = 1000
iterations = 1000
last_iterations = 100

r_values = np.linspace(r_min, r_max, n_r)

x = 0.1 * np.ones(n_r)

plt.figure(figsize=(14, 8))

for _ in range(iterations - last_iterations):
    x = variant_map(x, r_values)
    x[x > 10] = 10
    x[x < -10] = -10

print("Построение бифуркационной диаграммы...")
for _ in range(last_iterations):
    x = variant_map(x, r_values)
    plt.scatter(r_values, x, s=0.1, c='black', alpha=0.1)

plt.title(f'Бифуркационная диаграмма для Варианта N=3 (ISU 501983)')
plt.xlabel('Параметр r')
plt.ylabel('Установившиеся значения x_n')
plt.xlim(r_min, r_max)
plt.ylim(0, 1.5)
plt.grid(True, alpha=0.3)

plt.text(0.4, 0.1, 'Период 1', fontsize=12, color='blue')
plt.text(0.9, 0.1, 'Период 2', fontsize=12, color='blue')
plt.text(1.25, 0.1, 'Хаос', fontsize=12, color='red')

plt.show()
```
